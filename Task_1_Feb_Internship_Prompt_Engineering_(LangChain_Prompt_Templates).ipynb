2: Import Libraries

```python
from langchain_core.prompts import PromptTemplate, ChatPromptTemplate
```

Task 1 – Replace Hardcoded Prompt

```python
simple_explain_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} in simple terms for beginners"
)

def explain_topic(topic):
    return simple_explain_template.format(topic=topic)

# Test
print(explain_topic("Artificial Intelligence"))
```

Task 2 – Multi-Input Prompt

```python
multi_input_template = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template="Explain {topic} for {audience} in a {tone} tone"
)

def explain_for_audience(topic, audience, tone):
    return multi_input_template.format(
        topic=topic,
        audience=audience,
        tone=tone
    )

# Test Cases
print(explain_for_audience("AI", "beginners", "friendly"))
print(explain_for_audience("Python", "kids", "fun"))
print(explain_for_audience("Deep Learning", "engineers", "technical"))
```

Task 3 – Prompt Variations Engine

```python
teaching_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} clearly step by step"
)

interview_template = PromptTemplate(
    input_variables=["topic"],
    template="Ask 3 interview questions about {topic}"
)

storytelling_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} as a story"
)

def run_prompt_variations(topic):
    return {
        "Teaching": teaching_template.format(topic=topic),
        "Interview": interview_template.format(topic=topic),
        "Storytelling": storytelling_template.format(topic=topic)
    }

# Test
results = run_prompt_variations("Machine Learning")
for k, v in results.items():
    print(k, ":", v)
```

Task 4 – ChatPromptTemplate System

```python
def get_role_instruction(role):
    role_map = {
        "teacher": "You are a helpful teacher who explains concepts clearly.",
        "interviewer": "You are a professional interviewer asking smart questions.",
        "motivator": "You are a motivational speaker encouraging learning."
    }

    if role not in role_map:
        raise ValueError("Invalid role")

    return role_map[role]

def build_chat_prompt(role, topic):
    system_message = get_role_instruction(role)

    chat_template = ChatPromptTemplate.from_messages([
        ("system", system_message),
        ("user", "Explain {topic}")
    ])

    return chat_template.format_messages(topic=topic)

# Test
msgs = build_chat_prompt("teacher", "Neural Networks")
for m in msgs:
    print(m.type.upper(), ":", m.content)
```

Task 5 – Input Validation

```python
def validate_inputs(audience, tone):
    valid_audiences = ["beginner", "intermediate", "expert"]
    valid_tones = ["formal", "casual", "fun"]

    if audience not in valid_audiences:
        print("Invalid audience → defaulting to beginner")
        audience = "beginner"

    if tone not in valid_tones:
        print("Invalid tone → defaulting to formal")
        tone = "formal"

    return audience, tone

# Test
print(validate_inputs("kids", "playful"))
```

Task 6 – Prompt Generator App

```python
def generate_prompt(topic, audience, tone, style):
    audience, tone = validate_inputs(audience, tone)

    templates = {
        "teaching": PromptTemplate(
            input_variables=["topic", "audience", "tone"],
            template="Explain {topic} for {audience} in a {tone} teaching style"
        ),
        "interview": PromptTemplate(
            input_variables=["topic", "audience", "tone"],
            template="Create 3 interview questions about {topic} for {audience} in a {tone} style"
        ),
        "storytelling": PromptTemplate(
            input_variables=["topic", "audience", "tone"],
            template="Explain {topic} for {audience} in a {tone} storytelling style"
        )
    }

    return templates[style].format(
        topic=topic,
        audience=audience,
        tone=tone
    )

# Test
print(generate_prompt("Neural Networks", "beginner", "fun", "storytelling"))
```

Task 7 – Reusability Test

```python
reusable_template = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template="Explain {topic} for {audience} in a {tone} tone"
)

test_inputs = [
    ("AI", "beginner", "casual"),
    ("Python", "beginner", "fun"),
    ("Deep Learning", "expert", "formal"),
    ("Cybersecurity", "intermediate", "casual"),
    ("Data Science", "beginner", "formal")
]

for topic, audience, tone in test_inputs:
    print(reusable_template.format(
        topic=topic,
        audience=audience,
        tone=tone
    ))
```
